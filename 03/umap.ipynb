## t-SNEとUMAPを比較する

### MNISTデータを準備する

```python
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
digits.data.shape
```

```python
digits.target.shape
```

### t-SNEの場合

```python
from sklearn.manifold import TSNE

%timeit TSNE(n_components=2).fit_transform(digits.data)
```

```python
X_tsne = TSNE(n_components=2).fit_transform(digits.data)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=digits.target)
plt.colorbar()
```

```python
from umap import UMAP

%timeit UMAP(n_components=2).fit_transform(digits.data)
```

```python
X_umap = UMAP(n_components=2).fit_transform(digits.data)
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=digits.target)
plt.colorbar()
```
